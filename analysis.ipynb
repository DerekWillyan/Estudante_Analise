```python
dadosReprov = dados[dados['math score'] < 60]
```

```python
dadosReprov.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Supondo que 'dadosReprov' já está definido e é um DataFrame do Pandas

# Cria uma figura com tamanho personalizado
plt.figure(figsize=(13, 10))  # Ajuste o tamanho da figura conforme necessário

# Subplot 1: Gênero
plt.subplot(3, 2, 1)
dadosReprov.groupby('gender').size().plot.bar(color='gray')
plt.title('Gênero')

# Subplot 2: Raça/Etnia
plt.subplot(3, 2, 2)
dadosReprov.groupby('race/ethnicity').size().plot.bar(color='gray')
plt.title('Raça/Etnia')

# Subplot 3: Nível de educação dos pais
plt.subplot(3, 2, 3)
dadosReprov.groupby('parental level of education').size().plot.bar(color='gray')
plt.title('Educação dos Pais')

# Subplot 4: Tipo de lanche
plt.subplot(3, 2, 4)
dadosReprov.groupby('lunch').size().plot.bar(color='gray')
plt.title('Tipo de Lanche')

# Subplot 5: Curso de preparação para o teste
plt.subplot(3, 2, 5)
dadosReprov.groupby('test preparation course').size().plot.bar(color='gray')
plt.title('Preparação para o Teste')

# Ajusta o espaçamento entre os subplots
plt.subplots_adjust(wspace=0.4, hspace=1.0)  # Ajuste wspace e hspace conforme necessário

# Exibe a figura
plt.show()
```
